```python
import pandas as pd
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
import seaborn as sns

boston = load_boston()

df = pd.DataFrame(boston.data)
df.columns = boston.feature_names
df['PRICE'] = boston.target
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr());
```

```python
df.cov()
```

```python
X = df.loc[:, df.columns != 'PRICE']
y = df['PRICE']
```

```python
lreg = LinearRegression()
lreg.fit(X, y)
```

```python
df['PREDICTION'] = lreg.predict(X)
```

```python
df.head()
```

```python
df['ERROR'] = df['PRICE'] - df['PREDICTION']
```

```python
df['ERROR'].hist(bins=20)
```

```python
df['ERROR'].mean()
```

```python
sns.regplot(x='PREDICTION', y='ERROR', data=df);
```

```python
df.head()
```

```python
X = df.iloc[:, :13]
y = df['PRICE']
```

```python
lreg.fit(X, y)
```

```python
lreg.score(X, y)
```

```python
lreg.coef_
```

```python
coeffs = pd.DataFrame({
    'Variable': X.columns,
    'Weight'  : lreg.coef_
}).sort_values(by='Weight', ascending=False)
```
