```python
def f(row):
    if row['Sex'] == 'female' and row['Pclass'] == 1:
        val = 'F-High'
    if row['Sex'] == 'female' and row['Pclass'] in [2,3]:
        val = 'F-Low'
    if row['Sex'] == 'male' and row['Pclass'] == 1:
        val = 'M-High'
    if row['Sex'] == 'male' and row['Pclass'] in [2,3]:
        val = 'M-Low'
    return val
```

```python
df['Gender_Status'] = df.apply(f, axis = 1)
```

```python
def over_75(val):
    if val > 75:
        return 'Getting Up There'
    else:
        return 'Bloom of Youth'
```

```python
def get_status(val):
    if val.Survived == 1 and val.Sex == 'female':
        return 'Female-Survived'
    elif val.Survived == 1 and val.Sex == 'male':
        return 'Male-Survived'
    elif val.Survived == 0 and val.Sex == 'female':
        return 'Female-Died'
    elif val.Survived == 0 and val.Sex == 'male':
        return 'Male-Survived'
    else:
        return 'Other'
```

```python
%timeit df['Status'] = df.apply(get_status, axis=1)
```
